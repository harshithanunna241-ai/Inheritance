```python
class Company:
    def __init__(self, company_name, company_city, **kwargs):
        self.company_name = company_name
        self.company_city = company_city
        super().__init__(**kwargs)

class HR(Company):
    def __init__(self, Hr_Manager, **kwargs):
        self.Hr_Manager = Hr_Manager
        super().__init__(**kwargs)

class IT(Company):
    def __init__(self, programming_language, **kwargs):
        self.programming_language = programming_language
        super().__init__(**kwargs)

class Employee(HR, IT):
    def __init__(self, Employee_name, Hr_Manager,
                 company_name, company_city,
                 programming_language):

        super().__init__(
            Hr_Manager=Hr_Manager,
            programming_language=programming_language,
            company_name=company_name,
            company_city=company_city
        )

        self.Employee_name = Employee_name

    def display_all_details(self):
        print("Company Name:", self.company_name)
        print("Company City:", self.company_city)
        print("HR Manager:", self.Hr_Manager)
        print("Programming Language:", self.programming_language)
        print("Employee Name:", self.Employee_name)

child3 = Employee(
    "Harshitha",
    "Anupama",
    "Infosys",
    "Bangalore",
    "AI"
)

child3.display_all_details()
```

```python
class Hospital:
    def __init__(self, hospital_name, location, **kwargs):
        self.hospital_name = hospital_name
        self.location = location
        super().__init__(**kwargs)

class Doctor(Hospital):
    def __init__(self, doctor_name, **kwargs):
        self.doctor_name = doctor_name
        super().__init__(**kwargs)

class Nurse(Hospital):
    def __init__(self, nurse_name, **kwargs):
        self.nurse_name = nurse_name
        super().__init__(**kwargs)

class Patient(Doctor, Nurse):
    def __init__(self, patient_name, hospital_name, location, doctor_name, nurse_name):
        super().__init__(
            hospital_name=hospital_name,
            location=location,
            doctor_name=doctor_name,
            nurse_name=nurse_name
        )
        self.patient_name = patient_name

    def display_all_details(self):
        print("Hospital Name:", self.hospital_name)
        print("Location:", self.location)
        print("Doctor Name:", self.doctor_name)
        print("Nurse Name:", self.nurse_name)
        print("Patient Name:", self.patient_name)

child3 = Patient(
    "Harshitha",
    "Gandhi Hospital",
    "Secunderabad",
    "Dr.Anupama",
    "Pavani"
)

child3.display_all_details()
```

```python
class University:
    def __init__(self, university_name, university_city, **kwargs):
        self.university_name = university_name
        self.university_city = university_city
        super().__init__(**kwargs)

class College(University):
    def __init__(self, college_name, **kwargs):
        self.college_name = college_name
        super().__init__(**kwargs)

class School(University):
    def __init__(self, school_name, **kwargs):
        self.school_name = school_name
        super().__init__(**kwargs)

class Student(College, School):
    def __init__(self, student_name, university_name, university_city, college_name, school_name):
        super().__init__(
            university_name= university_name,
            university_city = university_city,
            college_name=college_name,
            school_name=school_name
        )
        self.student_name = student_name

    def display_all_details(self):
        print("University Name:", self.university_name)
        print("University City:", self.university_city)
        print("College Name:", self.college_name)
        print("School Name:", self.school_name)
        print("Student Name:", self.student_name)

child3 = Student(
    "JNTUK",
    "Kakinada",
    "Vignan Nirula",
    "Vignan",
    "Harshitha"
)

child3.display_all_details()
```

```python
class Bank:
    def __init__(self, bank_name, branch, **kwargs):
        self.bank_name = bank_name
        self.branch = branch
        super().__init__(**kwargs)

class Savings(Bank):
    def __init__(self, savings_acc_no, **kwargs):
        self.savings_acc_no = savings_acc_no
        super().__init__(**kwargs)

class Loan(Bank):
    def __init__(self, loan_amount, **kwargs):
        self.loan_amount = loan_amount
        super().__init__(**kwargs)

class Customer(Savings, Loan):
    def __init__(self, customer_name, bank_name, branch, savings_acc_no, loan_amount):
        super().__init__(
            bank_name = bank_name,
            branch = branch,
            savings_acc_no = savings_acc_no,
            loan_amount = loan_amount
        )
        self.customer_name = customer_name

    def display_all_details(self):
        print("Bank Name:", self.bank_name)
        print("Branch:", self.branch)
        print("Savings acc no:", self.savings_acc_no)
        print("Loan Amount:", self.loan_amount)
        print("Customer Name:", self.customer_name)

child3 = Customer(
    "SBI",
    "Ameerpet",
    "2345678",
    100000,
    "Harshitha"
)

child3.display_all_details()
```

```python
class Vehicle:
    def __init__(self, brand, model, **kwargs):
        self.brand = brand
        self.model= model
        super().__init__(**kwargs)

class Car(Vehicle):
    def __init__(self, car_type, **kwargs):
        self.car_type = car_type
        super().__init__(**kwargs)

class Bike(Vehicle):
    def __init__(self, bike_type, **kwargs):
        self.bike_type = bike_type
        super().__init__(**kwargs)

class Showroom(Car, Bike):
    def __init__(self, owner_name, brand, model, car_type,bike_type):
        super().__init__(
            brand= brand,
            model = model,
            car_type=car_type,
            bike_type=bike_type
        )
        self.owner_name = owner_name

    def display_all_details(self):
        print("Brand:", self.brand)
        print("Model:", self.model)
        print("Car Type:", self.car_type)
        print("Bike Type:", self.bike_type)
        print("Owner Name:", self.owner_name)

child3 = Showroom(
    "Honda",
    "Shine",
    "Swift",
    "Shine",
    "Harshitha"
)

child3.display_all_details()
```

```python
class School:
    def __init__(self, school_name, city, **kwargs):
        self.school_name = school_name
        self.city = city
        super().__init__(**kwargs)

class Teacher(School):
    def __init__(self, subject, **kwargs):
        self.subject = subject
        super().__init__(**kwargs)

class Library(School):
    def __init__(self, book_count, **kwargs):
        self.book_count = book_count
        super().__init__(**kwargs)

class Student(Teacher, Library):
    def __init__(self, student_name, school_name,city, subject, book_count):
        super().__init__(
            school_name= school_name,
            city = city,
            subject = subject,
            book_count = book_count
        )
        self.student_name = student_name

    def display_all_details(self):
        print("School Name:", self.school_name)
        print("City:", self.city)
        print("Subject:", self.subject)
        print("Book Count:", self.book_count)
        print("Student Name:", self.student_name)

child3 = Student(
    "Dr.K.K.R Gowtham",
    "Tenali",
    "Telugu",
    "10",
    "Harshitha"
)

child3.display_all_details()
```

```python
class ECommerce:
    def __init__(self, website_name, country, **kwargs):
        self.website_name = website_name
        self.country = country
        super().__init__(**kwargs)

class Product(ECommerce):
    def __init__(self, product_name, **kwargs):
        self.product_name = product_name
        super().__init__(**kwargs)

class Seller(ECommerce):
    def __init__(self, seller_name, **kwargs):
        self.seller_name = seller_name
        super().__init__(**kwargs)

class Customer(Product, Seller):
    def __init__(self, customer_name, website_name, country, product_name, seller_name):
        super().__init__(
            website_name= website_name,
            country = country,
            product_name=product_name,
            seller_name=seller_name
        )
        self.customer_name = customer_name

    def display_all_details(self):
        print("Website Name:", self.website_name)
        print("country:", self.country)
        print("Product Name:", self.product_name)
        print("Seller Name:", self.seller_name)
        print("Customer Name:", self.customer_name)

child3 = Customer(
    "Youtube",
    "USA",
    "Youtube music",
    "google",
    "Harshitha"
)

child3.display_all_details()
```

```python
from abc import ABC, abstractmethod

class Bank(ABC):
    def __init__(self, bank_name, branch, **kwargs):
        self.bank_name = bank_name
        self.branch = branch
        super().__init__(**kwargs)

    @abstractmethod
    def show_bank_details(self):
        pass


class ATM(Bank):
    def __init__(self, atm_id, **kwargs):
        self.atm_id = atm_id
        super().__init__(**kwargs)


class MobileBanking(Bank):
    def __init__(self, mobile_number, **kwargs):
        self.mobile_number = mobile_number
        super().__init__(**kwargs)


class User(ATM, MobileBanking):
    def __init__(self, user_name, bank_name, branch, atm_id, mobile_number):
        super().__init__(
            bank_name=bank_name,
            branch=branch,
            atm_id=atm_id,
            mobile_number=mobile_number
        )
        self.user_name = user_name

    def show_bank_details(self):
        print("Bank Name:", self.bank_name)
        print("Branch:", self.branch)
        print("ATM ID:", self.atm_id)
        print("Mobile Number:", self.mobile_number)
        print("User Name:", self.user_name)


user1 = User(
    "Harshitha",
    "SBI",
    "Tenali",
    "ATM123",
    "9876543210"
)

user1.show_bank_details()
```

```python
from abc import ABC, abstractmethod

class University(ABC):
    def __init__(self, university_name, city, **kwargs):
        self.university_name = university_name
        self.city = city
        super().__init__(**kwargs)

    @abstractmethod
    def show_university(self):
        pass


class College(University):
    def __init__(self, college_name, **kwargs):
        self.college_name = college_name
        super().__init__(**kwargs)


class Hostel(University):
    def __init__(self, hostel_name, **kwargs):
        self.hostel_name = hostel_name
        super().__init__(**kwargs)


class Student(College, Hostel):
    def __init__(self, student_name, university_name,
                 city, college_name, hostel_name):

        super().__init__(
            university_name=university_name,
            city=city,
            college_name=college_name,
            hostel_name=hostel_name
        )

        self.student_name = student_name

    def show_university(self):
        print("University Name:", self.university_name)
        print("City:", self.city)
        print("College Name:", self.college_name)
        print("Hostel Name:", self.hostel_name)
        print("Student Name:", self.student_name)


student1 = Student(
    "Harshitha",
    "JNTU",
    "Hyderabad",
    "JNTU College",
    "Girls Hostel"
)

student1.show_university()
```
